```python
eggnogdb_path = '/10Gdata/sjt/data/database/eggnog/eggnog-mapper-data/eggnog.db'

connect = sqlite3.connect(eggnogdb_path)
cursor = connect.cursor()
```

```python
cursor.execute("SELECT type,name,tbl_name,rootpage,sql FROM sqlite_schema;")
tuple_eggnogdb_sqlite_schema = cursor.fetchall()
df_eggnogdb_sqlite_schema = pd.DataFrame(tuple_eggnogdb_sqlite_schema,columns=['type','name','tbl_name','rootpage','sql'])
df_eggnogdb_sqlite_schema
```

```python
cursor.execute("SELECT * FROM og WHERE level=33090;")
tuple_eggnogdb_ogtable_where_level33090 = cursor.fetchall()
df_eggnogdb_ogtable_where_level33090 = pd.DataFrame(tuple_eggnogdb_ogtable_where_level33090,columns=['og','level','nm','description','COG_categories'])
df_eggnogdb_ogtable_where_level33090
```

```python
cursor.execute("SELECT * FROM event WHERE level=33090;")
tuple_eggnogdb_eventtable_where_level33090 = cursor.fetchall()
df_eggnogdb_eventtable_where_level33090 = pd.DataFrame(tuple_eggnogdb_eventtable_where_level33090,columns=['i','level','og','side1','side2'])
df_eggnogdb_eventtable_where_level33090
```

```python
cursor.execute("SELECT * FROM prots;")
tuple_eggnogdb_protstable = cursor.fetchall()
df_eggnogdb_protstable = pd.DataFrame(
    tuple_eggnogdb_protstable,
    columns=[
        'name',
        'bigg_reaction',
        'gos',
        'pfam',
        'pname',
        'ogs',
        'orthoindex',
        'kegg_ko',
        'kegg_cog',
        'kegg_disease',
        'kegg_ec',
        'kegg_brite',
        'kegg_rclass',
        'kegg_tc',
        'kegg_cazy',
        'kegg_pathway',
        'kegg_module',
        'kegg_reaction',
        'kegg_go',
        'kegg_drug',
        'kegg_pubmed',
        'kegg_network'
        ])
df_eggnogdb_protstable
```

```python
word = '@33090'

df_eggnogdb_protstable.fillna('',inplace=True)

m_right_startswith = df_eggnogdb_protstable['ogs'].apply(lambda x : x.startswith('%s,'%word))
m_right_endswith = df_eggnogdb_protstable['ogs'].apply(lambda x : x.endswith('%s'%word))
m_right_stuckinmiddle = df_eggnogdb_protstable['ogs'].apply(lambda x : x.find('%s,'%word)) != -1

m = m_right_startswith | m_right_endswith | m_right_stuckinmiddle

df_eggnogdb_protstable_where_level33090 = df_eggnogdb_protstable[m]
df_eggnogdb_protstable_where_level33090
```
